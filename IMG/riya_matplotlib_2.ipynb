```python
import matplotlib.pyplot as plt
x1 = [1, 2, 3, 4, 5]
y1 = [5, 10, 15, 20, 25]

plt.plot(x1, y1)
plt.title("Increasing Graph")
plt.show()
x2 = [1, 2, 3, 4, 5]
y2 = [25, 20, 15, 10, 5]

plt.plot(x2, y2)
plt.title("Decreasing Graph")
plt.show()
x3 = [1, 2, 3, 4, 5]
y3 = [10, 20, 5, 25, 15]

plt.plot(x3, y3)
plt.title("Up Down Up Graph")
plt.show()
x4 = [10, 20, 30, 40, 50]
y4 = [5, 10, 15, 20, 25]

plt.plot(x4, y4)
plt.title("Changed X Values")
plt.show()
x5 = [1, 2, 3, 4, 5]
y5 = [50, 30, 40, 20, 10]

plt.plot(x5, y5)
plt.title("Changed Y Values")
plt.show()

```

```python
import matplotlib.pyplot as plt
students = ["A", "B", "C", "D", "E"]
marks = [75, 80, 65, 90, 85]

plt.plot(students, marks)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [1000, 1500, 1200, 1800, 2000]

plt.plot(months, sales)
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temperature = [30, 32, 31, 35, 36, 34, 33]

plt.plot(days, temperature)
plt.title("Weekly Temperature")
plt.xlabel("Days")
plt.ylabel("Temperature (°C)")
plt.show()
x = [1, 2, 3, 4, 5]
y = [10, 30, 15, 40, 20]

plt.plot(x, y)
plt.title("Highest and Lowest Points")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()

print("Highest Point =", max(y))
print("Lowest Point =", min(y))
```
