```python
import matplotlib.pyplot as plt

# Data
products = ['Laptop', 'Phone', 'Tablet']
sales = [100, 150, 80]

# 1. Simple Bar Chart
plt.figure(figsize=(5,3))
plt.bar(products, sales)
plt.title("Product Sales")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()

# 2. Green Bar Chart with Grid
plt.figure(figsize=(5,3))
plt.bar(products, sales, color='green')
plt.title("Product Sales")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
students = ['Aman', 'Riya', 'Karan', 'Neha']
marks = [85, 92, 78, 88]

plt.figure(figsize=(5,3))
plt.bar(students, marks, color='skyblue')
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()

print("Highest Scoring Student =", students[marks.index(max(marks))])
print("Lowest Scoring Student =", students[marks.index(min(marks))])
plt.figure(figsize=(5,3))
plt.bar(products, sales, color=['red', 'blue', 'orange'])
plt.title("Product Sales with Different Colors")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1, 2, 3. Sales Comparison (2024 vs 2025)
products = ['Laptop', 'Phone', 'Tablet']
sales_2024 = [100, 150, 80]
sales_2025 = [120, 180, 100]

x = np.arange(len(products))
width = 0.35

plt.bar(x - width/2, sales_2024, width, color='blue', label='2024')
plt.bar(x + width/2, sales_2025, width, color='green', label='2025')

plt.xticks(x, products)
plt.title("Sales Comparison")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.legend()
plt.show()

# Product improved the most
increase = [sales_2025[i] - sales_2024[i] for i in range(len(products))]
print("Most Improved Product =", products[increase.index(max(increase))])

# 4. Marks Comparison (2024 vs 2025)
students = ['Aman', 'Riya', 'Karan']
marks_2024 = [75, 85, 80]
marks_2025 = [82, 90, 88]

x = np.arange(len(students))

plt.bar(x - width/2, marks_2024, width, color='skyblue', label='2024')
plt.bar(x + width/2, marks_2025, width, color='pink', label='2025')

plt.xticks(x, students)
plt.title("Marks Comparison")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sales Data
products = ['Laptop', 'Phone', 'Tablet']
sales_2024 = [100, 150, 80]
sales_2025 = [120, 180, 100]

x = np.arange(len(products))
w = 0.35

# Grouped Bar Chart - Sales
plt.bar(x-w/2, sales_2024, w, color='blue', label='2024')
plt.bar(x+w/2, sales_2025, w, color='green', label='2025')

plt.xticks(x, products)
plt.title("Sales Comparison 2024 vs 2025")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.legend()
plt.show()

print("Most Improved Product = Phone")

students = ['Aman', 'Riya', 'Karan']
marks_2024 = [75, 85, 80]
marks_2025 = [82, 90, 88]

x = np.arange(len(students))
plt.bar(x-w/2, marks_2024, w, color='orange', label='2024')
plt.bar(x+w/2, marks_2025, w, color='purple', label='2025')

plt.xticks(x, students)
plt.title("Marks Comparison 2024 vs 2025")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.legend()
plt.show()
```
