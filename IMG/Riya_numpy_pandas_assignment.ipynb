```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("Europe Sales Records.csv")

# 1. Missing Values
print("Missing Values in Each Column:")
print(df.isnull().sum())

# 2. Duplicate Rows
duplicate_count = df.duplicated().sum()
print("\nNumber of Duplicate Rows:", duplicate_count)

# 3. Remove Duplicates
df = df.drop_duplicates()

print("\nDataset Shape After Removing Duplicates:")
print(df.shape)
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("Europe Sales Records.csv")
country_revenue = df.groupby("Country")["Total Revenue"].sum()

print("Total Revenue Generated by Each Country:")
print(country_revenue)
top_10_countries = country_revenue.sort_values(ascending=False).head(10)

print("\nTop 10 Countries with Highest Revenue:")
print(top_10_countries)
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("Europe Sales Records.csv")
country_profit = df.groupby("Country")["Total Profit"].sum()

print("Total Profit for Each Country:")
print(country_profit)
top_5_profit = country_profit.sort_values(ascending=False).head(5)

print("\nTop 5 Countries with Highest Profit:")
print(top_5_profit)
bottom_5_profit = country_profit.sort_values().head(5)

print("\nBottom 5 Countries with Lowest Profit:")
print(bottom_5_profit)
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("Europe Sales Records.csv")
units_sold = df.groupby("Item Type")["Units Sold"].sum()

print("Total Units Sold for Each Item Type:")
print(units_sold)
item_revenue = df.groupby("Item Type")["Total Revenue"].sum()

print("\nItem Type with Highest Revenue:")
print(item_revenue.idxmax())
print("Revenue:", item_revenue.max())
item_profit = df.groupby("Item Type")["Total Profit"].sum()

print("\nItem Type with Highest Profit:")
print(item_profit.idxmax())
print("Profit:", item_profit.max())
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("Europe Sales Records.csv")
revenue_by_channel = df.groupby("Sales Channel")["Total Revenue"].sum()

print("Total Revenue by Sales Channel:")
print(revenue_by_channel)
profit_by_channel = df.groupby("Sales Channel")["Total Profit"].sum()

print("\nTotal Profit by Sales Channel:")
print(profit_by_channel)
channel_summary = df.groupby("Sales Channel")[["Total Revenue", "Total Profit"]].sum()

print("\nSales Channel Summary:")
print(channel_summary)

best_channel = channel_summary["Total Profit"].idxmax()

print("\nBest Performing Channel:", best_channel)
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("Europe Sales Records.csv")
order_count = df["Order Priority"].value_counts()

print("Count of Orders for Each Priority Level:")
print(order_count)
avg_profit = df.groupby("Order Priority")["Total Profit"].mean()

print("\nAverage Profit per Order Priority:")
print(avg_profit)
highest_priority = avg_profit.idxmax()
highest_profit = avg_profit.max()

print("\nPriority Level with Highest Average Profit:")
print(highest_priority)
print("Average Profit:", highest_profit)
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("Europe Sales Records.csv")
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Month"] = df["Order Date"].dt.to_period("M")
monthly_revenue = df.groupby("Month")["Total Revenue"].sum()

print("Total Revenue Generated Each Month:")
print(monthly_revenue)
print("\nMonth with Highest Revenue:")
print(monthly_revenue.idxmax())
print("Revenue:", monthly_revenue.max())
print("\nMonth with Lowest Revenue:")
print(monthly_revenue.idxmin())
print("Revenue:", monthly_revenue.min())
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("Europe Sales Records.csv")
total_revenue = df["Total Revenue"].sum()
total_profit = df["Total Profit"].sum()
best_country = (
    df.groupby("Country")["Total Profit"]
      .sum()
      .idxmax()
)
best_item_type = (
    df.groupby("Item Type")["Units Sold"]
      .sum()
      .idxmax()
)
best_sales_channel = (
    df.groupby("Sales Channel")["Total Profit"]
      .sum()
      .idxmax()
)
total_orders = df["Order ID"].count()
print("========== BUSINESS SUMMARY REPORT ==========")
print("Total Revenue          :", total_revenue)
print("Total Profit           :", total_profit)
print("Best Performing Country:", best_country)
print("Best Selling Item Type :", best_item_type)
print("Best Sales Channel     :", best_sales_channel)
print("Total Orders Processed :", total_orders)
```
