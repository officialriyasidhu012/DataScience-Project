```python
from google.colab import files

uploaded = files.upload()
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

```python
df = pd.read_csv('fresher_hiring_india_dataset.csv')

df.head()
```

```python
print(df.columns)
```

```python
#Gender Distribution Using Bar Chart
gender_count = df['gender'].value_counts()

plt.figure(figsize=(6,4))
plt.bar(gender_count.index, gender_count.values)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of Candidates')
plt.show()
```

```python
#Age Distribution Using Histogram
plt.figure(figsize=(8,5))
plt.hist(df['age'], bins=10)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Show the number of candidates in each degree
degree_count = df['degree'].value_counts()

plt.bar(degree_count.index, degree_count.values)
plt.title("Candidates in Each Degree")
plt.xticks(rotation=45)
plt.show()
```

```python
#Create a pie chart for hiring stages.
stage_count = df['hiring_stage'].value_counts()

plt.pie(stage_count.values,
        labels=stage_count.index,
        autopct='%1.1f%%')
plt.title("Hiring Stages")
plt.show()
```

```python
#Display top 10 job roles applied by candidates.
roles = df['job_role'].value_counts().head(10)

plt.bar(roles.index, roles.values)
plt.xticks(rotation=45)
plt.title("Top 10 Job Roles")
plt.show()
```

```python
#Show candidate count according to graduation year.
year_count = df['graduation_year'].value_counts().sort_index()

plt.plot(year_count.index,
         year_count.values,
         marker='o')

plt.title("Graduation Year Wise Candidates")
plt.xlabel("Graduation Year")
plt.ylabel("Count")
plt.show()
```

```python
#Display internship experience distribution.
internship = df['prior_internship'].value_counts()

plt.bar(internship.index,
        internship.values)

plt.title("Internship Experience")
plt.show()
```

```python
#Create a scatter plot between CGPA and salary.
salary_df = df.dropna(subset=['offered_salary_inr'])

plt.scatter(salary_df['cgpa'],
            salary_df['offered_salary_inr'])

plt.xlabel("CGPA")
plt.ylabel("Salary")
plt.title("CGPA vs Salary")
plt.show()
```

```python
#Display average salary by sector.
salary = df.groupby('sector')['offered_salary_inr'].mean()

plt.bar(salary.index,
        salary.values)

plt.xticks(rotation=90)
plt.title("Average Salary by Sector")
plt.show()
```

```python
#Show work type distribution (Remote, Hybrid, Onsite).
work = df['work_type'].value_counts()

plt.bar(work.index, work.values)

plt.title("Work Type Distribution")
plt.xlabel("Work Type")
plt.ylabel("Number of Candidates")

plt.show()
```

```python
locations = df['job_location'].value_counts().head(10)

plt.figure(figsize=(10,5))
plt.bar(locations.index, locations.values)

plt.title("Top 10 Job Locations")
plt.xlabel("Location")
plt.ylabel("Number of Candidates")

plt.xticks(rotation=45)
plt.show()
```

```python
#Show average projects count by hiring stage.
projects = df.groupby('hiring_stage')['projects_count'].mean()

plt.figure(figsize=(8,5))
plt.bar(projects.index, projects.values)

plt.title("Average Projects Count by Hiring Stage")
plt.xlabel("Hiring Stage")
plt.ylabel("Average Projects")

plt.xticks(rotation=45)
plt.show()
```

```python
#Compare selected and rejected candidates.
stage = df['hiring_stage'].value_counts()

plt.figure(figsize=(8,5))
plt.bar(stage.index, stage.values)

plt.title("Hiring Stage Distribution")
plt.xlabel("Hiring Stage")
plt.ylabel("Number of Candidates")

plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("fresher_hiring_india_dataset.csv")
```

```python
sns.countplot(data=df,
              x='hiring_stage',
              hue='gender')

plt.xticks(rotation=45)
plt.title("Hiring Stage by Gender")
plt.show()
```

```python
sns.boxplot(data=df,
            x='degree',
            y='cgpa')

plt.xticks(rotation=45)
plt.title("CGPA Distribution by Degree")
plt.show()
```

```python
sns.countplot(data=df,
              x='prior_internship',
              hue='gender')

plt.title("Internship Experience by Gender")
plt.show()
```

```python
sns.boxplot(data=df,
            x='degree',
            y='certifications_count')

plt.xticks(rotation=45)
plt.title("Certifications by Degree")
plt.show()
```

```python
sns.scatterplot(data=df,
                x='projects_count',
                y='cgpa')

plt.title("Projects Count vs CGPA")
plt.show()
```

```python
salary_df = df.dropna(subset=['offered_salary_inr'])

sns.scatterplot(data=salary_df,
                x='linkedin_connections',
                y='offered_salary_inr')

plt.title("LinkedIn Connections vs Salary")
plt.show()
```

```python
top_roles = df['job_role'].value_counts().head(10).index

sns.countplot(data=df,
              y='job_role',
              order=top_roles)

plt.title("Top 10 Job Roles")
plt.show()
```

```python
sns.countplot(data=df,
              x='sector',
              hue='hiring_stage')

plt.xticks(rotation=90)
plt.title("Sector Wise Hiring Stage")
plt.show()
```

```python
sns.countplot(data=df,
              x='referral_applied',
              hue='gender')

plt.title("Referral Applications by Gender")
plt.show()
```

```python
sns.boxplot(data=df,
            x='hiring_stage',
            y='cgpa')

plt.xticks(rotation=45)
plt.title("CGPA by Hiring Stage")
plt.show()
```

```python
num_df = df.select_dtypes(include=['int64','float64'])

plt.figure(figsize=(10,8))

sns.heatmap(num_df.corr(),
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Heatmap")
plt.show()
```
