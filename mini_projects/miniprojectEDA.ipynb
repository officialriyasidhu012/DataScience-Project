```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. Load Dataset
df = pd.read_csv("NetFlix.csv")
df.head()
```

```python
# 3. Dataset Overview
print("First 5 Rows:")
display(df.head())
print("Last 5 Rows:")
display(df.tail())
print("Shape of Dataset:")
print(df.shape)
print("\nDataset Information:")
df.info()
print("\nStatistical Summary:")
display(df.describe())
print("\nColumn Names:")
print(df.columns)
print("\nData Types:")
print(df.dtypes)
```

```python
# 4. Data Cleaning

# Create a copy of original dataset
df_clean = df.copy()

# Check missing values
print("Missing Values Before Cleaning:")
print(df_clean.isnull().sum())

# Fill missing values
df_clean['director'] = df_clean['director'].fillna('Unknown')
df_clean['cast'] = df_clean['cast'].fillna('Unknown')
df_clean['country'] = df_clean['country'].fillna('Unknown')
df_clean['rating'] = df_clean['rating'].fillna('Not Rated')

# Convert date_added column into date format
df_clean['date_added'] = pd.to_datetime(
    df_clean['date_added'],
    errors='coerce'
)

# Check duplicate rows
print("\nDuplicate Rows Before Removing:")
print(df_clean.duplicated().sum())

# Remove duplicate rows
df_clean = df_clean.drop_duplicates()

# Check results after cleaning
print("\nMissing Values After Cleaning:")
print(df_clean.isnull().sum())

print("\nDuplicate Rows After Removing:")
print(df_clean.duplicated().sum())

print("\nData Types After Cleaning:")
print(df_clean.dtypes)
```

```python
# 5. Outlier Detection
# Boxplot of release year
plt.figure(figsize=(8, 4))
sns.boxplot(x=df_clean['release_year'])
plt.title("Boxplot of Release Year")
plt.show()
```

```python
# IQR Method
Q1 = df_clean['release_year'].quantile(0.25)
Q3 = df_clean['release_year'].quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
print("Lower Limit:", lower_limit)
print("Upper Limit:", upper_limit)
```

```python
# Find outliers
outliers = df_clean[
    (df_clean['release_year'] < lower_limit) |
    (df_clean['release_year'] > upper_limit)
]
print("Number of Outliers:", outliers.shape[0])
```

```python
# Remove outliers
df_clean = df_clean[
    (df_clean['release_year'] >= lower_limit) &
    (df_clean['release_year'] <= upper_limit)
]
```

```python
# Boxplot after handling outliers
plt.figure(figsize=(8, 4))
sns.boxplot(x=df_clean['release_year'])
plt.title("Boxplot of Release Year After Handling Outliers")
plt.show()
```

```python
# 7. Bivariate Analysis

# 1. Rating according to Content Type
plt.figure(figsize=(10, 5))
sns.countplot(
    data=df_clean,
    x='rating',
    hue='type'
)
plt.xticks(rotation=45)
plt.title("Rating According to Content Type")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
# 2. Movies and TV Shows by Release Year
plt.figure(figsize=(10, 5))
sns.histplot(
    data=df_clean,
    x='release_year',
    hue='type',
    bins=20,
    multiple='stack'
)
plt.title("Movies and TV Shows by Release Year")
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.show()
```

```python
# 8. Correlation Analysis

# First create content_age feature
# (Agar tumne feature engineering wala cell abhi nahi run kiya hai)
df_clean['content_age'] = 2026 - df_clean['release_year']
# Select only numerical columns
numeric_df = df_clean.select_dtypes(include=['int64', 'float64'])
# Show correlation values
print("Correlation Values:")
print(numeric_df.corr())
# Create heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 9. Feature Engineering

# Create a new column: content_age
# It tells us how old the movie or TV show is
df_clean['content_age'] = 2026 - df_clean['release_year']
# Show new column
df_clean[['title', 'release_year', 'content_age']].head()
```

```python
# Create one more useful feature: content_category
# Recent Content = released in 2015 or after
# Old Content = released before 2015

df_clean['content_category'] = np.where(
    df_clean['release_year'] >= 2015,
    'Recent Content',
    'Old Content'
)
# Show new feature
df_clean[['title', 'release_year', 'content_category']].head()
```

```python
# Graph for new feature
plt.figure(figsize=(6, 4))
sns.countplot(data=df_clean, x='content_category')
plt.title("Recent Content vs Old Content")
plt.xlabel("Content Category")
plt.ylabel("Count")
plt.show()
```
